```python
x2=['a','b','c','d','e']
y2=np.random.rand(5)
```

```python
x2
```

```python
y2
```

```python
# x vs y
plt.bar(x2,y2) #catagorical v/s numerical data
```

```python
plt.figure(figsize=(5,3))
plt.xlabel("catagorical data")
plt.ylabel("numerical data")
plt.title("BAR PLOT")
# x vs y
plt.bar(x2,y2,color='r')
#catagorical v/s numerical data
plt.grid()
```

```python
### for rotation use plt.barh()
plt.figure(figsize=(5,3))
plt.xlabel("catagorical data")
plt.ylabel("numerical data")
plt.title("BAR PLOT")
# x vs y
plt.barh(x2,y2,color='r')
#catagorical v/s numerical data
plt.grid()
```

```python
# 3-D

x4=np.random.rand(50)
y4=np.random.rand(50)
z=np.random.rand(50)
fig=plt.figure()
ax=fig.add_subplot(projection='3d')   # 3 D representation
ax.scatter(x4,y4,z,c='red',alpha=.8)
plt.xlabel('X Axis')
plt.ylabel('Y axis')
# z label not availabel
plt.zlabel("Z AXis")
plt.show
```
